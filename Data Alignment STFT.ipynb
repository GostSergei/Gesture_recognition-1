### Getting gesture start time

```python
def get_first_step_idx(sig):
    f, t, Zxx = signal.stft(signal.medfilt(sig,3),
                        fs=30,
                        nperseg=16,
                        noverlap=15,
                        boundary='constant')
    
    mean_spec = np.mean(np.abs(Zxx)[:1,], axis=0)
    mean_spec = mean_spec/np.max(mean_spec)

    diff_spec = np.diff(mean_spec,2)
    diff_spec = diff_spec/np.max(diff_spec)
    
    peaks2, _ = signal.find_peaks(diff_spec, prominence=0.3, distance=10)
    
    return peaks2[0]
```

```python
def sort_by(input_array, sort_values):
    output = [x for _, x in sorted(zip(sort_values, input_array),key=lambda pair: pair[0])]
    return np.array(output)

sort_by(input_array=[2,3,4], sort_values=[3,1,2])
```

#### getting indexes of first encountered step-change in the signal

```python
first_step_idx = []
last_step_idx = []
for sig in clean_data:
    first_idx = get_first_step_idx(sig)
    last_idx = get_first_step_idx(sig[::-1])
    
    first_step_idx.append(first_idx)
    last_step_idx.append(120-last_idx)    
```

### Cleaning and Aligning

```python
data_full = mt_file['data']
labels_full = mt_file['labels'][0]
data_full.shape
```

```python
# bad idxs
bad_idx_final = list(set(bad_idxs_init) | set(short_gest_idxs))
mask = np.ones(len(data_full))
mask[bad_idx_final] = 0
```

```python
clean_data_full = np.array(data_full)[mask==1]
clean_labels_full = np.array(labels_full)[mask==1]
clean_first_step_idx = np.array(first_step_idx)[mask==1]
clean_last_step_idx = np.array(last_step_idx)[mask==1]
```

```python
plt.figure(figsize=(6,2))
plt.hist(np.array(clean_last_step_idx) - np.array(clean_first_step_idx), bins=100)
plt.grid(True)
plt.show()
```

```python
Num = 1623
plt.figure(figsize=(6,2))
plt.plot(clean_data_full[Num,::,12,2], 'r')
ddd = clean_data_full[Num,::,12,2]
plt.scatter([clean_first_step_idx[Num], clean_last_step_idx[Num]], [np.mean(ddd),np.mean(ddd)], color='r', s=100)
plt.show()
```

```python
aligned_data_full = []
for step_idx, sig in zip(clean_first_step_idx, clean_data_full):
    new_sig = np.roll(sig, shift=20-step_idx, axis=0)
#     new_sig = shift(sig, shift=np.array([20-step_idx,0]), mode='reflect')
    aligned_data_full.append(new_sig)

aligned_data_full = np.array(aligned_data_full) 

clean_last_step_idx += 20 - clean_first_step_idx
clean_first_step_idx = np.ones_like(clean_last_step_idx)*20

aligned_data_full.shape
```
